```python
def increase(curr,interval):
    hour=int(curr[:2])
    minute=int(curr[2:4])
    second=int(curr[4:])
    hour_=int(interval[:2])
    minute_=int(interval[2:4])
    second_=int(interval[4:])
    hour=hour+hour_
    minute=minute+minute_
    second=second+second_
    if second>59:
        minute+=1
        second=second%60
    if minute>59:
        hour+=1
        minute=minute%60
    hour=hour%24
    
    hour=str(hour) if len(str(hour))==2 else '0'+str(hour)
    minute=str(minute) if len(str(minute))==2 else '0'+str(minute)
    second=str(second) if len(str(second))==2 else '0'+str(second)
    return hour+minute+second
    
    
def irrigation(plots,motors,starttime,endtime,motorruntime,irrigationcycle):
    plotqueue=[('D'+str(i),starttime) for i in range(1,plots+1)]
    motorsqueue=[('M'+str(i),starttime) for i in range(1,motors+1)]
    curr=starttime
    
    while(plotqueue and motorsqueue):
        if curr>=plotqueue[0][-1] and curr>=motorsqueue[0][-1]:
            motor,motortime=motorsqueue.pop(0)
            plot,plottime=plotqueue.pop(0)
            print('starttime :',motortime,'endtime :',increase(motortime,motorruntime),'motor :',motor,'plot :',plot)
            plotqueue.append((plot,increase(curr,irrigationcycle)))
            motorsqueue.append((motor,increase(curr,motorruntime)))
        curr=increase(curr,'000001')
        if curr>=endtime:
            return
```

```python
irrigation(4,2,'070000','190000','000500','002000')
```
